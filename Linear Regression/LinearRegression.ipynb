<h2>Linear Regression</h2>

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
data = pd.read_csv('Admission_Predict.csv')
```

```python
data.head()
```

```python
data.describe()
```

Checking Linearity among input data and target data using Scatterplot and Heatmap


Target column: Chance of Admit 

```python
sns.pairplot(data)
```

```python
data_corr = data.corr()
sns.heatmap(data_corr,annot=True)
```

We see that CGPA and Chance of Admit  have positive correlation = 0.87


Therefore we take Input column: CGPA

```python
from sklearn import linear_model as lm
regressor = lm.LinearRegression()

#entire column used for training
train_x = np.asanyarray(data[['CGPA']])
train_y = np.asanyarray(data[['Chance of Admit ']])

#using least squares in background
regressor.fit(train_x, train_y)
print('Coefficients: ', regressor.coef_)
print('Intercept: ', regressor.intercept_)
```

```python
print("The equation for linear regression is y = ",regressor.coef_[0][0],"x + ",regressor.intercept_[0])
```

Comparing the Predicted values and the Actual values by Plotting

```python
plt.scatter(data['CGPA'], data['Chance of Admit '], color='blue', label="Actual")
plt.plot(train_x, regressor.coef_[0][0]*train_x + regressor.intercept_[0], '-g',label="Predicted")
plt.title("Graduate Admission")
plt.legend()
plt.xlabel("CGPA")
plt.ylabel("Chance of Admit ")
plt.show()
```

Comparing y_pred and Chance of Admit side by side by inserting into the dataframe

```python
y_pred = regressor.predict(train_x)
data['Predicted'] = y_pred
```

```python
data.head()
```

Measuring the performance of the model using error metrics

```python
from sklearn import metrics
print("MAE: ", metrics.mean_absolute_error(train_y, y_pred))
print('MSE:', metrics.mean_squared_error(train_y, y_pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(train_y, y_pred)))
```

Using R2-score to check the variability between the linear regression model and the actual data

```python
from sklearn.metrics import r2_score

test_x = np.asanyarray(data[['CGPA']])
test_y = np.asanyarray(data[['Chance of Admit ']])
test_y_pred = regressor.predict(test_x)

print("MAE:" , np.mean(np.absolute(test_y_pred - test_y)))
print("MSE: " , np.mean((test_y_pred - test_y) ** 2))
print("R2-score: " , r2_score(test_y, test_y_pred) )
#if r2 score is higher, then variability is less, which means, the regression prediction fits the data well
```
